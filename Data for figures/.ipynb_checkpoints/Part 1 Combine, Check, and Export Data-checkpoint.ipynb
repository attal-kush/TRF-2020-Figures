```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

path = os.getcwd()
files = os.listdir(path)
print(files)

list_to_fill = [os.path.join(root, name)
             for root, dirs, files in os.walk(path)
             for name in files
             if name.endswith((".xlsx", ".xls")) & name.startswith(('Rat'))]
```

# Step 2: Check for Errors

### a) Find and Fix Consecutive Start or Starts within Each Behavior

- This code looks at the **Status** for each individual behavior and tries to find instances where there are **consecutive starts/stops**. 
    - For example, if a feeding row has a start and the next feeding row has a start, it will find the row and to which rat and hour that the first start was in. Then it is up to the student to decide which start is **vaild** or if they missed when the first feeding ended in the data and should **manually add the stop** for the first feeding behavior into the data.
    

- If this code finds a **start** as the very **last row** on the excel sheet then delete that instance


- This is from people not ending/**stop**ing an activity like when the rat initially falls asleep/resting and then recounting it being asleep/resting in the next hour interval. 
    - For example, the rat falls asleep at 1:59:58 and it is recorded. Then when the next person looks at the hour 2 footage, they would imediately say that the rat started falling asleep at 2:00:01, while in reality the rat fell asleep before the hour mark. Therefore, the row after the 2 hour mark should be deleted, because it is double counting and will mess up histograms of rat behavior.

```python
behaviors = ["Feeding", "Sucrose", "Sleeping/Resting", "Rearing", "Grooming", "Water"]
for i in list_to_fill:
    rat = pd.DataFrame()
    test = pd.read_excel(i, index_col = "seconds")
    rat = rat.append(test)

    for j in behaviors:
        filtbehav = rat[rat.Behavior == j]
        for i in range(len(filtbehav)):
            try:
                if filtbehav.iloc[i].Status == "START":
                    if filtbehav.iloc[i + 1].Status == "STOP":
                        pass
                    else:
                        print(filtbehav.iloc[i])
                elif filtbehav.iloc[i].Status == "STOP":
                    if filtbehav.iloc[i + 1].Status == "START":
                        pass
                    else:
                        print(filtbehav.iloc[i])
            except IndexError:
                pass
```

The information about the problem rows will contain info that is useful to find the rows. 
If you have no cells appearing, **congratulations**, then your Excel sheets are free of START/STOP erros
