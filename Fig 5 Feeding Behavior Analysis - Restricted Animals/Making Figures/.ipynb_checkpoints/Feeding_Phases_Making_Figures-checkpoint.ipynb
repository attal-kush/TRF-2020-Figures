# Behavioral Research: Does sugar and fat change the feeding behavior of animals?

This Jupyter notebook plots the **average** feeding activity of the rats in 3 different phases.

## Calculate total food consumption for each rat during light and dark phases

```python
def light_summary(df, timeseries):
    # Time of light on, the data start at light on
    light_start = pd.to_datetime('1970-01-02 09:00:00')
    #Separate dataframes into light and dark times
    dark = timeseries[timeseries.index<light_start]
    light = timeseries[timeseries.index>light_start]
    # Calculate total food consumption per rat
    total_eating_light = light.sum(axis=0).astype(int)
    total_eating_dark  = dark.sum(axis=0).astype(int)
    # Record food comsumption into a new dataframe, using rat number as index
    for x in total_eating_light.index:
        ind = int(x[-2:])
        df.loc[ind]=[total_eating_light[x], total_eating_dark[x]]
    return (df)
```

Compile food consumption in each experimental group into single dataframe

```python
column_names = ["light_food", "dark_food"]
total_feeding_data = pd.DataFrame(columns = column_names)

total_feeding_data = light_summary(total_feeding_data, cont_adlib_feeding)
total_feeding_data = light_summary(total_feeding_data, cont_restr_feeding)
total_feeding_data = light_summary(total_feeding_data, hfhs_adlib_feeding)
total_feeding_data = light_summary(total_feeding_data, hfhs_restr_feeding)

total_feeding_data.head()
```

```python
total_feeding_data['group']=groups_data.loc[total_feeding_data.index].diet+' '+groups_data.loc[total_feeding_data.index].feeding
total_feeding_data.head()
```

### Download plot parameters

```python
plot_data_location = "https://www.dropbox.com/s/djpbvky0mwxpups/plotting_by_group.csv?dl=1"
plot_parameters = pd.read_csv(plot_data_location, index_col=0)
plot_parameters
```

### Function which plots given variable of the dataframe in box plots

```python
def box_plot_graph(df, time, title, ymax = 2000):
    ax1 = sns.swarmplot(x=df.group, y=df[time], color='black', size=3)
    ax = sns.boxplot    (x=df.group, y=df[time], color='white', linewidth=1, palette=plot_parameters.fill_color, showfliers = False)
    # Make hatched pattern to distinguish between Ad lib and restriction
    for i, hatch, patch in zip(plot_parameters.hatch_colors, plot_parameters.hatches, ax.artists):
        patch.set_hatch(hatch)
        patch.set_edgecolor(i)
    # Add x- and y-labels, ticks, and units
    plt.ylabel(title, fontsize=10, fontweight = "bold", color = "black")
    plt.ylim(0,ymax)
    plt.xlabel('')
    plt.xticks(range(len(plot_parameters.index)), ('Cont\nAL', 'Cont\nRes', 'HFHS\nAL', 'HFHS\nRes'), fontweight = 'bold', fontsize=10, color = "black")
    plt.yticks(fontweight = 'bold', fontsize=10, color = "black")
    plt.setp(ax.lines, color="black")
    return()
```

### 1. Compare Light vs Day vs Total Feeding Activity in Rats

```python
plt.figure(figsize=(18, 5))
plt.subplot(1,3,1)
box_plot_graph(total_feeding_data, 'total_food', "Total Eating (sec)", 8000)
plt.subplot(1,3,3)
box_plot_graph(total_feeding_data, 'light_food', "Light Eating (sec)", 8000)
plt.subplot(1,3,2)
box_plot_graph(total_feeding_data, 'dark_food', "Dark Eating (sec)", 8000)
#plt.savefig('total_eating_time.png')
plt.show()
```

### Create Dataframe to: 
1. Calculate Hourly Feeding Activity in Rats (number of seconds spent feeding per hour)
2. Add 3 columns to calculate average feeding activity for each rat in (1) first 2 hours, (2) mid 4 hours, (3) last 2 hours for the 8-hour feeding window for all rats

```python
cont_adlib_feeding_hourly = cont_adlib_feeding.resample("1H",how=lambda x: x.values.sum()).T
hfhs_adlib_feeding_hourly = hfhs_adlib_feeding.resample("1H",how=lambda x: x.values.sum()).T
cont_restr_feeding_hourly = cont_restr_feeding.resample("1H",how=lambda x: x.values.sum()).T
hfhs_restr_feeding_hourly = hfhs_restr_feeding.resample("1H",how=lambda x: x.values.sum()).T

hourly_feeding_frames = [cont_adlib_feeding_hourly, cont_restr_feeding_hourly, hfhs_adlib_feeding_hourly, hfhs_restr_feeding_hourly]
feeding_hourly_frame = pd.concat(hourly_feeding_frames)
feeding_hourly_frame.index = feeding_hourly_frame.index.map(lambda x: int(str(x)[3:]))
feeding_hourly_frame['group']=groups_data.loc[feeding_hourly_frame.index].diet+' '+groups_data.loc[feeding_hourly_frame.index].feeding


feeding_hourly_frame.columns = ["Hour21", "Hour22", "Hour23", "Hour0", 
                                "Hour1", "Hour2", "Hour3", "Hour4",
                                "Hour5", "Hour6", "Hour7", "Hour8",
                                "Hour9", "Hour10", "Hour11", "Hour12",
                                "Hour13", "Hour14", "Hour15", "Hour16",
                                "Hour17", "Hour18", "Hour19", "Hour20",
                                "group"]
phase_frame = pd.DataFrame()
phase_frame["First 2-Hour Feeding Rate Per Hour"] = feeding_hourly_frame[["Hour23", "Hour0"]].mean(axis=1)
phase_frame["Mid 5-Hour Feeding Rate Per Hour"] = feeding_hourly_frame[["Hour1", "Hour2", "Hour3", "Hour4", "Hour5"]].mean(axis=1)
phase_frame["Last 1-Hour Feeding Rate Per Hour"] = feeding_hourly_frame[["Hour6"]].mean(axis=1)
phase_frame["group"] = feeding_hourly_frame["group"]

phase_frame
```

### 2. Compare 3-Phase Feeding Activity in Rats Using Random Hours

```python
plt.figure(figsize=(18, 5))
plt.subplot(1,3,1)
box_plot_graph(feeding_hourly_frame, 'Hour23', "First Hour Eating (sec)")
plt.subplot(1,3,2)
box_plot_graph(feeding_hourly_frame, 'Hour0', "2nd Hour Eating (sec)")
plt.subplot(1,3,3)
box_plot_graph(feeding_hourly_frame, 'Hour6', "Last Hour Eating (sec)")
#plt.savefig('total_eating_time.png')
plt.show()
```

## 3. Compare Average Rates at the 3 Phases of Night

```python
plt.figure(figsize=(18, 5))
plt.subplot(1,3,1)
box_plot_graph(phase_frame, "First 2-Hour Feeding Rate Per Hour", "First Phase Consumption Rate (sec/hour)")
plt.subplot(1,3,2)
box_plot_graph(phase_frame, "Mid 5-Hour Feeding Rate Per Hour", "Second Phase Consumption Rate (sec/hour)")
plt.subplot(1,3,3)
box_plot_graph(phase_frame, "Last 1-Hour Feeding Rate Per Hour", "Third Phase Consumption Rate (sec/hour)")
#plt.savefig('total_eating_time.png')
plt.show()
```

```python
#Create a new column that orders the values in the order given
phase_frame["barplot_order"] = pd.Categorical(phase_frame.group, categories=["control ad lib","HFHS ad lib","control restriction","HFHS restriction"], ordered=True)

# Reorder the dataframe based on custom order
barplot_frame = phase_frame.sort_values("barplot_order").melt(id_vars = ["group"], value_vars = 
                 ["First 2-Hour Feeding Rate Per Hour", "Mid 5-Hour Feeding Rate Per Hour", 
                  "Last 1-Hour Feeding Rate Per Hour"])
barplot_frame.head()
```

```python
# Create a custom plot parameter for the barplot figure
barplot_plot_parameters = plot_parameters.reindex(["control ad lib", "HFHS ad lib", "control restriction", "HFHS restriction"])

# Add custom edge colors
barplot_plot_parameters["edgecolors"] = ["black", "darkred", "grey", "red"]
barplot_plot_parameters
```

## Function that creates bar plots

```python
def adjust_lightness(color, amount=0.5):
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], max(0, min(1, amount * c[1])), c[2])
```

```python
def bar_plot_graph(df, title, ymax = 1000):
    ax = sns.barplot(x="group", y="value", data = df, hue = "variable", 
                     ci = 68, capsize=.1, errwidth=0.9)
    
    # Collect the color properties for the bars
    first_phase_bars = ax.patches[0:4]
    second_phase_bars = ax.patches[4:8]
    third_phase_bars = ax.patches[8:12]
    
    
    # Make hatched pattern to distinguish between Ad lib and restriction
    for i, hatch, fill, firstphase, secondphase, thirdphase,  in zip(barplot_plot_parameters.edgecolors, barplot_plot_parameters.hatches, barplot_plot_parameters.fill_color, first_phase_bars, second_phase_bars, third_phase_bars):
        # Set same face color for all 3 phases
        firstphase.set_facecolor(fill)
        secondphase.set_facecolor(fill)
        thirdphase.set_facecolor(fill)
        
        # Set same hatch color for all 3 phases
        firstphase.set_hatch(hatch)
        secondphase.set_hatch(hatch)
        thirdphase.set_hatch(hatch)
        
        # Set same edge color for all 3 phases - set new column in DropBox
        firstphase.set_edgecolor(i)
        secondphase.set_edgecolor(i)
        thirdphase.set_edgecolor(i)
        
        
    # Add x- and y-labels, ticks, and units
    plt.ylabel(title, fontsize=15, fontweight = "bold", color = "black")
    plt.ylim(0,ymax)
    plt.xlabel('')
    plt.xticks(range(len(barplot_plot_parameters.index)), ('Cont\nAL', 'HFHS\nAL', 'Cont\nRes', 'HFHS\nRes'), fontweight = 'bold', fontsize=20, color = "black")
    plt.yticks(fontweight = 'bold', fontsize=15, color = "black")
    
    # Set error bar colors
    plt.setp(ax.lines[0:3] + ax.lines[6:9] + ax.lines[12:15] + ax.lines[18:21] + ax.lines[24:27] + ax.lines[30:33], color = "black")
    plt.setp(ax.lines[3:6] + ax.lines[9:12] + ax.lines[15:18] + ax.lines[21:24] + ax.lines[27:30] + ax.lines[33:36], color = "darkred")
    
    # Remove legend
    ax.get_legend().remove()
    
    # Add Significance markers
    plt.annotate('ab', (1.65, 850), fontsize=15, color = 'black', fontweight='bold')
    plt.annotate('c', (3.22, 600), fontsize=15, color = 'black', fontweight='bold')
    
```
