# Metabolite Figures

In this jupyter notebook, I will create a 4 boxplot graphs, each showing the terminal level of 1 metabolite among the 4 diet groups. The colors will be gray for the Control rats and red for the HFHS rats. Rats on a TRF diet will have dashed lines. The 4 metabolites will be: 
1. Adiponectin
2. Leptin
3. Triglyceride

4. Cholesterol  <font color=red>Should we replace Cholesterole with GSP? </font> 

After plotting the 4 boxplots, I will perform statsitcal analysis of each of the metabolite levels to see which diet group significantly differed from another diet group

```python
#----------------------------------------------------------
# Import important libraries
#----------------------------------------------------------
import pandas  as pd
import numpy   as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import statsmodels.stats.multicomp
sns.set()

#Set the background and size of the figure
sns.set_style("whitegrid", {'axes.grid' : False, 'axes.edgecolor': 'black', 'font.family': 'Arial'})
my_pal = {"Control Ad Lib": "gray", "HFHS Ad Lib": "red", "Control Restricted": "white", "HFHS Restricted": "white"}
plt.rcParams['hatch.linewidth'] = 3
plt.rcParams['figure.dpi'] = 1000
```

```python
#----------------------------------------------------------
# Method to Create Plots
#----------------------------------------------------------
def making_plot(metabolite, unit, name):
    #Set the size, dashed lines, and colors for the figure
    plt.figure(figsize=(10, 11))
    ax1 = sns.swarmplot(x=master_data.index, y=metabolite, color='black', size=6)
    ax = sns.boxplot(x=master_data.index, y=metabolite, color='white', linewidth=1, palette=my_pal, showfliers = False)
    hatches = ["", "/", "", "/"]
    colors = ["black", "gray", "black", "red"]
    for i, hatch, patch in zip(colors, hatches, ax.artists):
        patch.set_hatch(hatch)
        patch.set_edgecolor(i)

    # Add x- and y-labels, ticks, and units
    plt.ylabel(name + ' (' + unit + ')', fontsize=25, fontweight = "bold", color = "black")
    plt.xlabel('')
    plt.xticks(fontweight = 'bold', fontsize=15, color = "black")
    plt.yticks(fontweight = 'bold', fontsize=20, color = "black")
    plt.setp(ax.lines, color="black")

    # Clean up and save the figure
    sns.despine()
    plt.tight_layout()
    
#----------------------------------------------------------
# Method to Run T-Test Stats
#----------------------------------------------------------
def running_stats(control_adlib_metabolite, control_restr_metabolite, hfhs_adlib_metabolite, hfhs_restr_metabolite, name):
    #---------------------------------------------------
    #   T-test for metabolite between Control Ad Lib and Control Restricted
    #---------------------------------------------------
    t, p = stats.ttest_ind(control_adlib_metabolite, control_restr_metabolite)
    print('p-value for ' + name + ' between Control Ad Lib and Control Restricted:', p)
    #---------------------------------------------------
    #   T-test for metabolite between Control Ad Lib and HFHS Ad Lib
    #---------------------------------------------------
    t, p = stats.ttest_ind(control_adlib_metabolite, hfhs_adlib_metabolite)
    print('p-value for ' + name + ' between Control Ad Lib and HFHS Ad Lib:', p)
    #---------------------------------------------------
    #   T-test for metabolite between Control Ad Lib and HFHS Restricted
    #---------------------------------------------------
    t, p = stats.ttest_ind(control_adlib_metabolite, hfhs_restr_metabolite)
    print('p-value for ' + name + ' between Control Ad Lib and HFHS Restricted:', p)
    #---------------------------------------------------
    #   T-test for metabolite between Control Restricted and HFHS Ad Lib
    #---------------------------------------------------
    t, p = stats.ttest_ind(control_restr_metabolite, hfhs_adlib_metabolite)
    print('p-value for ' + name + ' between Control Restricted and HFHS Ad Lib:', p)
    #---------------------------------------------------
    #   T-test for metabolite between Control Restricted and HFHS Restricted
    #---------------------------------------------------
    t, p = stats.ttest_ind(control_restr_metabolite, hfhs_restr_metabolite)
    print('p-value for ' + name + ' between Control Restricted and HFHS Restricted:', p)
    #---------------------------------------------------
    #   T-test for metabolite between HFHS Ad Lib and HFHS Restricted
    #---------------------------------------------------
    t, p = stats.ttest_ind(hfhs_adlib_metabolite, hfhs_restr_metabolite)
    print('p-value for ' + name + ' between HFHS Ad Lib and HFHS Restricted:', p)

#----------------------------------------------------------
# Method to run Anova
#----------------------------------------------------------
def anova_analysis(metabolite, anova_data):
    formula = metabolite + ' ~ C(diet) + C(feeding_schedule) + C(diet):C(feeding_schedule)'
    model = ols(formula, anova_data).fit()
    aov_table = anova_lm(model, typ=2)
    
    return aov_table
```

## Creating Boxplot Figures

<font color=red>We need to convert individual plots into functions. Then make 2 x 2 subplot figure and call appropriate function for each subplot. </font> 

```python
#----------------------------------------------------------
# Collect all the relevant data from the .xlxs file into one 'master_data' dataframe
#----------------------------------------------------------
master_data = pd.read_excel('2018VT_termination_data_master_document.xlsx', index_col='diet')
master_data['group'] = (master_data.index) + " " + (master_data.feeding_schedule)
master_data = master_data.set_index('group')
master_data = master_data.rename(columns={" Leptin": "Leptin", "triglyceride (mg/mL)": "Triglyceride"}, index = {"control ad lib": "Control Ad Lib", "control restriction": "Control Restricted", "HFHS ad lib": "HFHS Ad Lib", "HFHS restriction": "HFHS Restricted"})

#----------------------------------------------------------
# Collect all the relevant data from the .xlxs file into one 'master_data' dataframe
#----------------------------------------------------------
anova_data = pd.read_excel('2018VT_termination_data_master_document.xlsx', index_col='Rat')
anova_data = anova_data.rename(columns={" Leptin": "Leptin", "triglyceride (mg/mL)": "Triglyceride"}, index = {"control ad lib": "Control Ad Lib", "control restriction": "Control Restricted", "HFHS ad lib": "HFHS Ad Lib", "HFHS restriction": "HFHS Restricted"})
```

```python
#----------------------------------------------------------
# Create the plots and run ANOVA stats
#----------------------------------------------------------

# ADIPONECTIN
making_plot(master_data.Adiponectin, "mcg/mL", "Adiponectin")
plt.annotate('###', (0.9, 9.5), fontsize=20, color = 'black', fontweight='bold')
plt.annotate('*', (1.96, 9.5), fontsize=30, color = 'black', fontweight='bold')
plt.annotate('**', (2.93, 7.5), fontsize=30, color = 'black', fontweight='bold')
plt.savefig('Adiponectin Levels at Termination.tif')
plt.show()

# ANOVA
print(anova_analysis('Adiponectin', anova_data))

# Tukey post-hoc
mc = statsmodels.stats.multicomp.MultiComparison(anova_data['Leptin'], anova_data['feeding_schedule'])
mc_results = mc.tukeyhsd()
print(mc_results)
```

```python
# LEPTIN
making_plot(master_data.Leptin, "mcg/mL", "Leptin")
plt.annotate('##', (2.93, 2.1), fontsize=20, color = 'black', fontweight='bold')
plt.savefig('Leptin Levels at Termination.tif')
plt.show()

# ANOVA
print(anova_analysis('Leptin', anova_data))
```

```python
# TRIGLYCERIDE
making_plot(master_data.Triglyceride, "mg/mL", "Triglyceride")
plt.annotate('***', (1.87, 5), fontsize=30, color = 'black', fontweight='bold')
plt.annotate('**', (2.92, 3.3), fontsize=30, color = 'black', fontweight='bold')
plt.annotate('#', (2.96, 3.6), fontsize=20, color = 'black', fontweight='bold')
plt.savefig('Triglyceride Levels at Termination.tif')
plt.show()

# ANOVA
print(anova_analysis('Triglyceride', anova_data))
```

```python
# CHOLESTEROL
making_plot(master_data.Cholesterol, "mg/dL", "Cholesterol")
plt.annotate('***', (1.87, 65), fontsize=30, color = 'black', fontweight='bold')
plt.annotate('***', (2.87, 48), fontsize=30, color = 'black', fontweight='bold')
plt.savefig('Cholesterol Levels at Termination.tif')
plt.show()

# ANOVA
print(anova_analysis('Cholesterol', anova_data))
```

# Perform T-test statistical analysis of each of the 4 metabolite levels

I will use Student's t-test (default version in scipy) to compare metabolite levels between diet groups

```python
#----------------------------------------------------------
#Created the 4 new data frames for all 4 diet groups
#----------------------------------------------------------
Control_adlib = master_data.index == 'Control Ad Lib'
HFHS_adlib = master_data.index == 'HFHS Ad Lib'
Control_adlib  = master_data[Control_adlib].set_index('Rat')
HFHS_adlib = master_data[HFHS_adlib].set_index('Rat')

Control_restr = master_data.index == 'Control Restricted'
HFHS_restr = master_data.index == 'HFHS Restricted'
Control_restr  = master_data[Control_restr].set_index('Rat')
HFHS_restr = master_data[HFHS_restr].set_index('Rat')

HFHS_restr
```

```python
running_stats(Control_adlib.Leptin, Control_restr.Leptin, HFHS_adlib.Leptin, HFHS_restr.Leptin, "Leptin")
```

```python
running_stats(Control_adlib.Adiponectin, Control_restr.Adiponectin, HFHS_adlib.Adiponectin, HFHS_restr.Adiponectin, "Adiponectin")
```

# Cholesterol Analysis

```python
running_stats(Control_adlib.Cholesterol, Control_restr.Cholesterol, HFHS_adlib.Cholesterol, HFHS_restr.Cholesterol, "Cholesterol")
```

# Triglyceride Analysis

```python
running_stats(Control_adlib.Triglyceride, Control_restr.Triglyceride, HFHS_adlib.Triglyceride, HFHS_restr.Triglyceride, "Triglyceride")
```

# Panel for Metabolites

```python
#----------------------------------------------------------
# Method to Create Panel of Metabolites
#----------------------------------------------------------
def making_panel(metabolite, unit, name):
    #Set the size, dashed lines, and colors for the figure
    ax1 = sns.swarmplot(x=master_data.index, y=metabolite, color='black', size=2)
    ax = sns.boxplot(x=master_data.index, y=metabolite, color='white', linewidth=1, palette=my_pal, showfliers = False)
    hatches = ["", "/", "", "/"]
    colors = ["black", "gray", "black", "red"]
    for i, hatch, patch in zip(colors, hatches, ax.artists):
        patch.set_hatch(hatch)
        patch.set_edgecolor(i)

    # Add x- and y-labels, ticks, and units
    plt.ylabel(name + ' (' + unit + ')', fontsize=10, fontweight = "bold", color = "black")
    plt.xlabel('')
    plt.xticks(fontweight = 'bold', fontsize=6, color = "black")
    plt.yticks(fontweight = 'bold', fontsize=10, color = "black")
    plt.setp(ax.lines, color="black")

    # Clean up and save the figure
    sns.despine()
    plt.tight_layout()

plt.figure(figsize = (7.48, 10))
# CHOLESTEROL
plt.subplot(2, 2, 1)
making_panel(master_data.Cholesterol, "mg/dL", "Cholesterol")
plt.annotate('***', (1.9, 65), fontsize=10, color = 'black', fontweight='bold')
plt.annotate('***', (2.9, 48), fontsize=10, color = 'black', fontweight='bold')
plt.figtext(0.005, 0.98, "A", fontsize = 15, color = "black", fontweight = "bold")

# TRIGLYCERIDE
plt.subplot(2, 2, 2)
making_panel(master_data.Triglyceride, "mg/mL", "Triglyceride")
plt.annotate('***', (1.9, 5), fontsize=10, color = 'black', fontweight='bold')
plt.annotate('**', (2.95, 3.3), fontsize=10, color = 'black', fontweight='bold')
plt.annotate('#', (2.95, 3.5), fontsize=10, color = 'black', fontweight='bold')
plt.figtext(0.5, 0.98, "B", fontsize = 15, color = "black", fontweight = "bold")

# LEPTIN
plt.subplot(2, 2, 3)
making_panel(master_data.Leptin, "mcg/mL", "Leptin")
plt.annotate('##', (2.9, 2.1), fontsize=10, color = 'black', fontweight='bold')
plt.figtext(0.005, 0.49, "C", fontsize = 15, color = "black", fontweight = "bold")

# ADIPONECTIN
plt.subplot(2, 2, 4)
making_panel(master_data.Adiponectin, "mcg/mL", "Adiponectin")
plt.annotate('###', (0.85, 9.5), fontsize=10, color = 'black', fontweight='bold')
plt.annotate('*', (1.97, 9.5), fontsize=10, color = 'black', fontweight='bold')
plt.annotate('**', (2.95, 7.5), fontsize=10, color = 'black', fontweight='bold')
plt.figtext(0.5, 0.49, "D", fontsize = 15, color = "black", fontweight = "bold")

plt.savefig("Metabolite Panel.tif")
plt.show()
```
