# Plotting timeline of weight and final adiposity

Download timeseries for weight and terminate data from the Dropbox/../Figures/Data for Figures. Plot first pannel (A) weigt vs. time and the second pannel (B) final adiposity for all 4 experimental groups

```python
#----------------------------------------------------------
# Import important libraries
#----------------------------------------------------------
import pandas as pd
import numpy as np
import datetime
import os 
import matplotlib.pyplot as plt
import seaborn as sns
import math
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import statsmodels.stats.multicomp
import matplotlib.dates as mdates
sns.set()

#Set the background and size of the figure
sns.set_style("whitegrid", {'axes.grid' : False, 'axes.edgecolor': 'black', 'font.family': 'Arial'})
plt.rcParams['hatch.linewidth'] = 3
plt.rcParams['figure.dpi'] = 1000
```

# Body Weight

```python
weight_data_location = "https://www.dropbox.com/s/wktthzuhv7vgvox/2018VT%20-%20final%20weight%20log.csv?dl=1"
body_weight = pd.read_csv(weight_data_location).T
body_weight.columns = body_weight.iloc[0]

metafile = body_weight.iloc[1:3].T
body_weight.drop(['Rat', 'Diet', 'Feeding'], inplace = True)
body_weight

# Changing column names for body_weight to include day numbers
#label = ['diet', 'feeding_schedule']
#for i in range(55):
#    label.append('Day' + str(i + 1))
#body_weight.index = label OR body_weight.columns = label
```

## Download parmeters for plotting each group

```python
plot_data_location = "https://www.dropbox.com/s/5ev83xuz7ujuane/plotting_by_group.csv?dl=1"
plot_parameters = pd.read_csv(plot_data_location, index_col=0)
plot_parameters
```

## Loop through Diet and feeding conditions and plot timeline for each group

```python
def plot_weight_vs_time(metafile, body_weight):
    # Separate into subgroups according to diet and food accesibility
    regime = metafile.Feeding.unique()
    diet   = metafile.Diet.unique()
    days = np.arange(0, body_weight.shape[0])
    restriction_start = 27
    for x in diet:
        for y in regime:
            group = str(x)+' '+str(y)
            ids = metafile[(metafile.Diet==x) & (metafile.Feeding==y)].index
            weight_series = body_weight[ids]
            weight_series.to_csv(group+'.csv', index=True)
            plt.errorbar(days, weight_series.mean(axis = 1), yerr=weight_series.std(axis = 1)/np.sqrt(weight_series.shape[1]), color = plot_parameters.loc[group].color, marker = plot_parameters.loc[group].marker, ls=plot_parameters.loc[group].line_type, markevery=5, markersize = 5, capsize=5, errorevery=5, lw = 1)
            plt.axvline(x=restriction_start, ymin=0, ymax=400, color='black', ls ='--')
    plt.xlabel('Days of on a diet', fontsize=10, fontweight = "bold", color = "black")
    plt.ylabel('Body Weight (g)', fontsize=10, fontweight = "bold", color = "black")
    plt.yticks(fontweight = 'bold', fontsize=10, color = "black")
    plt.xticks(fontweight = 'bold', fontsize=10, color = "black")
```

## Download terminal data master document

```python
master_data_location = 'https://www.dropbox.com/s/eu0szifsb9yimo3/2018VT_termination_data_master_document.csv?dl=1'
master_data = pd.read_csv(master_data_location, index_col=0)
# Correct column names
master_data = master_data.rename(columns={" Leptin": "Leptin", "triglyceride (mg/mL)": "Triglyceride"})
master_data['group']=master_data.diet+' '+master_data.feeding_schedule
master_data.head()
```

## Make function to plot final adipocity by group

```python
def make_fat_graph(master_data, plot_parameters):
    ax1 = sns.swarmplot(x=master_data.group, y=master_data.total_fat_pad, color='black', size=2)
    ax = sns.boxplot    (x=master_data.group, y=master_data.total_fat_pad, color='white', linewidth=1, palette=plot_parameters.fill_color, showfliers = False)
    # Make hatched pattern to distinguish between Ad lib and restriction
    for i, hatch, patch in zip(plot_parameters.hatch_colors, plot_parameters.hatches, ax.artists):
        patch.set_hatch(hatch)
        patch.set_edgecolor(i)

    # Add x- and y-labels, ticks, and units
    plt.ylabel('Total Fat Mass (g)', fontsize=12, fontweight = "bold", color = "black")
    plt.xlabel('')
    plt.xticks(fontweight = 'bold', fontsize=6, color = "black")
    plt.yticks(fontweight = 'bold', fontsize=10, color = "black")
    plt.setp(ax.lines, color="black")
    plt.annotate('*', (1.98, 26.8), fontsize=10, color = 'black', fontweight='bold')
    plt.annotate('**', (2.95, 20), fontsize=10, color = 'black', fontweight='bold')
    ax.set_xticklabels(('Cont\n\nAd Lib', 'Cont\n\nRestr', 'HFHS\n\nAd Lib', 'HFHS\n\nRestr'),ha='center', fontsize=12)
    # hide tick lines for x axis
    ax.tick_params(axis='x', which='both',length=0)  
```

## Make final 2-pannel figure for weight and fat

```python
plt.figure(figsize = (7.48, 3.5))
plt.subplot(1,2,1)
plot_weight_vs_time(metafile, body_weight)
plt.text(-7.5, 380, 'A', fontsize=15, fontweight='bold')
plt.subplot(1,2,2)
make_fat_graph(master_data, plot_parameters)
plt.text(-0.8, 28, 'B', fontsize=15, fontweight='bold')
# Clean up and save the figure
sns.despine()
plt.tight_layout()
plt.savefig('Fig2_weight_and_fat.tif', dpi = 1000)
plt.show()
  
```

# Statistics Testing for Body Weight

```python
#----------------------------------------------------------
#Created the 4 new data frames for all 4 diet groups
#----------------------------------------------------------
Control_adlib_bw  = pd.read_csv('control ad lib.csv', index_col=0)
Control_restr_bw  = pd.read_csv('control restriction.csv', index_col=0)
HFHS_adlib_bw     = pd.read_csv('HFHS ad lib.csv', index_col=0)
HFHS_restr_bw     = pd.read_csv('HFHS restriction.csv', index_col=0)

#----------------------------------------------------------
#Created BW dataframe appropriate for ANOVA testing
#----------------------------------------------------------
body_weight_anova = pd.read_csv(weight_data_location, index_col = 0)
```

```python
#---------------------------------------------------
#   T-test for total_fat_pad between Control Ad Lib and HFHS Ad Lib
#---------------------------------------------------
for day in body_weight.index:
    t, p = stats.ttest_ind(Control_adlib_bw.loc[day], HFHS_adlib_bw.loc[day])
    if p < 0.05:
        print('p-value for body weight of ' + day + ' between Control Ad Lib and HFHS Ad Lib:', p)
```

```python
#---------------------------------------------------
#   T-test for total_fat_pad between Control Restricted and HFHS Restricted
#---------------------------------------------------
for day in body_weight.index:
    t, p = stats.ttest_ind(Control_restr_bw.loc[day], HFHS_restr_bw.loc[day])
    if p < 0.05:
        print('p-value for body weight of ' + day + ' between Control Restriction and HFHS Restriction:', p)
```

```python
#---------------------------------------------------
#   T-test for total_fat_pad between Control Ad Lib and Control Restricted
#---------------------------------------------------
ca_vs_cr_days = []
for day in body_weight.index:
    t, p = stats.ttest_ind(Control_adlib_bw.loc[day], Control_restr_bw.loc[day])
    if p < 0.05:
        print('p-value for body weight of ' + day + ' between Control Ad Lib and Control Restricted:', p)
        ca_vs_cr_days.append(day)
#---------------------------------------------------
#   T-test for total_fat_pad between HFHS Ad Lib and HFHS Restricted
#---------------------------------------------------
ha_vs_hr_days = []
for day in body_weight.index:
    t, p = stats.ttest_ind(HFHS_adlib_bw.loc[day], HFHS_restr_bw.loc[day])
    if p < 0.05:
        print('p-value for body weight of ' + day + ' between HFHS Ad Lib and HFHS Restricted:', p)
        ha_vs_hr_days.append(day)
```

```python
#---------------------------------------------------
#   ANOVA Analysis for Days of Significance
#---------------------------------------------------
for day in ca_vs_cr_days:
    print(day)
    formula = 'Q("' + day + '") ~ C(Diet) + C(Feeding) + C(Diet):C(Feeding)'
    model = ols(formula, body_weight_anova).fit()
    aov_table = anova_lm(model, typ=2)
    print(aov_table)
```

# Perform statistical analysis of fat pad mass

## Statistical analysis has already been run for body weight. It is located in the source Excel file for body weight.

I will use Student's t-test (default version in scipy) to compare fat pad mass between diet groups


```python
#----------------------------------------------------------
#Created the 4 new data frames for all 4 diet groups
#----------------------------------------------------------
Control_adlib = (master_data.diet == 'control') & (master_data.feeding_schedule == "ad lib")
HFHS_adlib = (master_data.diet == 'HFHS') & (master_data.feeding_schedule == "ad lib")
Control_adlib  = master_data[Control_adlib]
HFHS_adlib = master_data[HFHS_adlib]

Control_restr = (master_data.diet == 'control') & (master_data.feeding_schedule == "restriction")
HFHS_restr = (master_data.diet == 'HFHS') & (master_data.feeding_schedule == "restriction")
Control_restr  = master_data[Control_restr]
HFHS_restr = master_data[HFHS_restr]
```

```python
#---------------------------------------------------
#   T-test for total_fat_pad between Control Ad Lib and Control Restricted
#---------------------------------------------------
t, p = stats.ttest_ind(Control_adlib.total_fat_pad, Control_restr.total_fat_pad)
print('p-value for total_fat_pad between Control Ad Lib and Control Restricted:', p)
#---------------------------------------------------
#   T-test for total_fat_pad between Control Ad Lib and HFHS Ad Lib
#---------------------------------------------------
t, p = stats.ttest_ind(Control_adlib.total_fat_pad, HFHS_adlib.total_fat_pad)
print('p-value for total_fat_pad between Control Ad Lib and HFHS Ad Lib:', p)
#---------------------------------------------------
#   T-test for total_fat_pad between Control Ad Lib and HFHS Restricted
#---------------------------------------------------
t, p = stats.ttest_ind(Control_adlib.total_fat_pad, HFHS_restr.total_fat_pad)
print('p-value for total_fat_pad between Control Ad Lib and HFHS Restricted:', p)
#---------------------------------------------------
#   T-test for total_fat_pad between Control Restricted and HFHS Ad Lib
#---------------------------------------------------
t, p = stats.ttest_ind(Control_restr.total_fat_pad, HFHS_adlib.total_fat_pad)
print('p-value for total_fat_pad between Control Restricted and HFHS Ad Lib:', p)
#---------------------------------------------------
#   T-test for total_fat_pad between Control Restricted and HFHS Restricted
#---------------------------------------------------
t, p = stats.ttest_ind(Control_restr.total_fat_pad, HFHS_restr.total_fat_pad)
print('p-value for total_fat_pad between Control Restricted and HFHS Restricted:', p)
#---------------------------------------------------
#   T-test for total_fat_pad between HFHS Ad Lib and HFHS Restricted
#---------------------------------------------------
t, p = stats.ttest_ind(HFHS_adlib.total_fat_pad, HFHS_restr.total_fat_pad)
print('p-value for total_fat_pad between HFHS Ad Lib and HFHS Restricted:', p)
```
